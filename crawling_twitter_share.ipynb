```python
def list_Twt(driver) :    
    TimeOut = 10

    # 게시글 추출
    #스크롤 내리기 이동 전 위치
    scroll_location = driver.execute_script("return document.body.scrollHeight")
    cnt = 0
    htmls = []
    while True:
        
        #현재 스크롤의 가장 아래로 내림
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

        #전체 스크롤이 늘어날 때까지 대기
        time.sleep(7)

        #늘어난 스크롤 높이
        scroll_height = driver.execute_script("return document.body.scrollHeight")
#         if cnt == 2 : 
#             break

        #늘어난 스크롤 위치와 이동 전 위치 같으면(더 이상 스크롤이 늘어나지 않으면) 종료
        if scroll_location == scroll_height:
            break

        #같지 않으면 스크롤 위치 값을 수정하여 같아질 때까지 반복
        else:
            #스크롤 위치값을 수정
            htmls.append(driver.page_source)
            scroll_location = driver.execute_script("return document.body.scrollHeight")
#             cnt += 1
    
    return htmls

#[출처] Twitter 과거 데이터 크롤링|작성자 Pisces
```

```python
def parse_Twt(htmls) : 
    prs_html = []
    for html in htmls : 
        soup = BeautifulSoup(html,"html.parser")   # 리스트 내 html 요소 parsing
        div_find = soup.select("div.css-901oao.r-18jsvk2.r-37j5jr.r-a023e6.r-16dba41.r-rjixqe.r-bcqeeo.r-bnwqim.r-qvutc0")
        for idx, div in enumerate(div_find) : 
            prs_html.append(div.text)
    return prs_html
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.tokenize import word_tokenize
import re
import csv
import pandas as pd

def save_pre_Twt(txt_list, file_name=None) : 
    strip_whitespc = [string.strip() for string in txt_list]   # 공백 제거 
    spl_txt = [string.split() for string in strip_whitespc]   # whitespace 기준으로 split
    rmv_id = [re.sub("r@[a-zA-Z0-9]+"," ",string)for string in strip_whitespc]   # 트위터 id 부분 제거
    leave_abc = [re.sub(r"[^a-zA-z]"," ",string) for string in rmv_id]    # 영어 이외 전부 제거
    upper_txt = [string.upper() for string in leave_abc]     # 대문자화
#     return upper_txt
    s = pd.Series(upper_txt)    # pandas로 변환
    s.to_csv(file_name)     # csv로 전환

    return
```
